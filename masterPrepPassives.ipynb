```python
os.chdir("/Users/nduran/Dropbox (ASU)/ScottProject/activePassive/dataAnalysis2/all_mt_files/nns/")
# os.chdir("/Users/nduran/Dropbox (ASU)/ScottProject/activePassive/dataAnalysis2/missing_3_NS/")

NS_SUB = 1
NNS_SUB = 100

allfiles = sort(os.listdir('.'))
for f, file1 in enumerate(allfiles):
   
#     if file1.endswith("1740.txt"): 
    if file1.endswith(".txt"): 
    
        os.chdir("/Users/nduran/Dropbox (ASU)/ScottProject/activePassive/dataAnalysis2/all_mt_files/nns/")
#         os.chdir("/Users/nduran/Dropbox (ASU)/ScottProject/activePassive/dataAnalysis2/missing_3_NS/")
        
        f1 = open(file1,"r")
        allTrials = f1.readlines()
        subjnum = file1.strip(".txt").split("_")[0]
        print subjnum
        
        allI = []
        allX = []
        allY = []
        allT = []
        
        for j, line1 in enumerate(allTrials):
            if "RAW TRACKS (X coordinates)" in line1:
                startX = j
            if "RAW TRACKS (Y coordinates)" in line1:
                startY = j
            if "RAW TRACKS (time)" in line1: 
                startT = j
#         print startX, startY, startT
        
### GET TRIAL INFO AND X COORDINATES

        ## loop through all trials again at appropriate start and stop times to find when the coordinates begin     
        for k in range(startX,startY-1,1):        
            trialInfo = []
            xInfo = []
            trial = allTrials[k]
            trial = trial.replace(';',',').split(',')  
            ## loop within each trial 
            evalStart = 0
            for i in range(0, len(trial), 1):  
                if trial[i] == "hover":
                    evalStart = i
                    break             
                        
            ## loop WITHIN each trial again at appropriate start and stop times to get coordinates and trial info             
            if len(trial) > 18:
                for kk in range(0, evalStart-1, 1): 
                    trialInfo.append(trial[kk])  
                for kk in range(evalStart+1, len(trial), 1): 
                    xInfo.append(trial[kk])           
        
            allI.append(trialInfo)
            allX.append(xInfo)
            
### GET Y COORDINATES

        ## loop through all trials again at appropriate start and stop times to find when the coordinates begin     
        for k in range(startY,startT-1,1):        
            yInfo = []
            trial = allTrials[k]
            trial = trial.replace(';',',').split(',')      
            ## loop within each trial 
            evalStart = 0
            for i in range(0, len(trial), 1):                    
                if trial[i] == "hover":
                    evalStart = i
                    break 
                    
            ## loop WITHIN each trial again at appropriate start and stop times to get coordinates and trial info             
            if len(trial) > 18:
                for kk in range(evalStart+1, len(trial), 1): 
                    yInfo.append(trial[kk])                  
            
            allY.append(yInfo)
              
### GET TIME 
    
        ## loop through all trials again at appropriate start and stop times to find when the coordinates begin     
        for k in range(startT,len(allTrials),1):        
            tInfo = []
            trial = allTrials[k]
            trial = trial.replace(';',',').split(',')        
            ## loop within each trial 
            evalStart = 0
            for i in range(0, len(trial), 1):                    
                if trial[i] == "hover":
                    evalStart = i
                    break 

            ## loop WITHIN each trial again at appropriate start and stop times to get coordinates and trial info             
            if len(trial) > 18:
                for kk in range(evalStart+1, len(trial), 1): 
                    tInfo.append(trial[kk])                  
            
            allT.append(tInfo)
        
        if len(allI) == len(allX) == len(allY) == len(allT):
            
            os.chdir("/Users/nduran/Dropbox (ASU)/ScottProject/activePassive/dataAnalysis2/mt_transformed/nns_clean/")
#             os.chdir("/Users/nduran/Dropbox (ASU)/ScottProject/activePassive/dataAnalysis2/mt_transformed/ns_reject/")
            
            ## get each trial
            counter = 1
            for rr in range(0, len(allI), 1):
                if len(allI[rr]) > 0:
                    
                    newname = str(NNS_SUB) + "_" + str(counter) + ".txt"
                    newfile = open(newname,"w")

                    # the first line
                    newfile.write(allT[rr][1] + "\t" + allX[rr][1] + "\t" + allY[rr][1] + "\t" + allI[rr][0] + "\t" + allI[rr][1] + "\t" + allI[rr][2] + "\t" + allI[rr][3] + "\t" + allI[rr][4] + "\t" + allI[rr][5] + "\t" + allI[rr][6] + "\t" + allI[rr][7] + "\t" + allI[rr][8] + "\t" + allI[rr][9] + "\t" + allI[rr][10] + "\t" + allI[rr][11] + "\n")
                    # everything past the first line
                    for nn in range(2,len(allX[rr])-1,1): # everything past the first line
                        newfile.write(allT[rr][nn] + "\t" + allX[rr][nn] + "\t" + allY[rr][nn] + "\t?\t?\t?\t?\t?\t?\t?\t?\t?\t?\t?\t?\n")

                    newfile.close()
                    
                    counter += 1
    
        else:
            print "*******bad*********"

    NS_SUB += 1
    NNS_SUB += 1
```
